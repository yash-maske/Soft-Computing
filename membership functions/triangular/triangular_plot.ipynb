```python
import matplotlib.pyplot as plt

def plot_from_file(filename, label):
    x, y = [], []
    with open(filename, "r") as f:
        for line in f:
            xi, yi = map(float, line.split())
            x.append(xi)
            y.append(yi)
    plt.plot(x, y, label=label)

# Plot all
plot_from_file("triangular_a_0.txt", "Triangular_a_0")
plot_from_file("triangular_a_1.txt", "triangular_a_1")
plot_from_file("triangular_a_2.txt", "triangular_a_2")
plot_from_file("triangular_a_3.txt", "triangular_a_3")
plot_from_file("triangular_a_4.txt", "triangular_a_4")

plt.legend()
plt.xlabel("x")
plt.ylabel("Membership value (x)")
plt.title("Fuzzy Membership Functions")
plt.grid(True)
plt.show()
```
