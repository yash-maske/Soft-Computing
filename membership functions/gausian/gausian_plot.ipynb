```python
import matplotlib.pyplot as plt

def plot_from_file(filename, label):
    x, y = [], []
    with open(filename, "r") as f:
        for line in f:
            xi, yi = map(float, line.split())
            x.append(xi)
            y.append(yi)
    plt.plot(x, y, label=label)

plot_from_file("gaussian_mean_3.txt", "Gaussian_mean_3")
plot_from_file("gaussian_mean_4.txt", "Gaussian_mean_4")
plot_from_file("gaussian_mean_5.txt", "Gaussian_mean_5")
plot_from_file("gaussian_mean_6.txt", "Gaussian_mean_6")
plot_from_file("gaussian_mean_7.txt", "Gaussian_mean_7")




plt.legend()
plt.xlabel("x")
plt.ylabel("Membership value (x)")
plt.title("Fuzzy Membership Functions")
plt.grid(True)
plt.show()
```
