```python
import matplotlib.pyplot as plt

def plot_from_file(filename, label):
    x, y = [], []
    with open(filename, "r") as f:
        for line in f:
            xi, yi = map(float, line.split())
            x.append(xi)
            y.append(yi)
    plt.plot(x, y, label=label)

# Plot all

plot_from_file("sigmoidal_a_1.txt", "sigmoidal_a_1")
plot_from_file("sigmoidal_a_2.txt", "sigmoidal_a_2")
plot_from_file("sigmoidal_a_3.txt", "sigmoidalad_3")
plot_from_file("sigmoidal_a_4.txt", "sigmoidalad_4")
plot_from_file("sigmoidal_a_5.txt", "sigmoidalad_5")

plt.legend()
plt.xlabel("x")
plt.ylabel("Membership value (x)")
plt.title("Fuzzy Membership Functions")
plt.grid(True)
plt.show()
```
